## Exc 3.3

```python
vars = ['tid=*', 'civilstand=U']   

def construct_link(x, y):
    start_url = 'https://api.statbank.dk/v1/data/'
    final_url = start_url + x + '/JSONSTAT?' + y[0] + '&' + y[1]
    print(final_url)
    
construct_link('FOLK1A', vars)
```

```python
vars2 = ['tid=*', 'barnkon=P']
construct_link('FOD', vars2)
```

```python
url2 = 'https://api.statbank.dk/v1/data/FOD/JSONSTAT?tid=*&barnkon=P'
response2 = requests.get(url2)
response2.ok
```

```python
jsonresponse2 = response2.json()
jsonresponse2
```

```python
import pandas as pd
data = pd.DataFrame.from_dict(jsonresponse2, orient="columns")
data.values
```

## Exc 4 pandas

```python
import pandas as pd 
data = pd.read_csv('https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/by_year/1864.csv.gz',
                  compression = 'gzip', header=None)
mydf = pd.DataFrame(data.iloc[:,0:4])
```

```python
mydf.columns = ['station identifier', 'data', 'observation type', 'obs values']
mydf
```

```python
newdf = mydf.loc[(mydf['station identifier'] == 'ITE00100550') & (mydf['observation type'] == 'TMAX')]
newdf
```

```python
tmax_F = newdf['obs values']*1.8 + 32
newdf = newdf.assign(tmax_F=tmax_F)
```

```python
newdf = newdf.reset_index(drop=True)
newdf
```

```python
newdf = newdf.sort_values(by='obs values')


print(newdf.head(1))
print(newdf.tail(1))
```

```python
newdf.to_csv('temperatuur.csv', index=False)
```
